```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define the data
X = [2, 5, 8, 15, 20, 17, 18]
Y = [3, 4, 16, 18, 21, 10, 12]

# Create a DataFrame from the data
data = pd.DataFrame({'X': X, 'Y': Y})

# Scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatterplot')
plt.show()

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Boxplot')
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
plt.hist(data['X'], bins='auto', alpha=0.7, label='X')
plt.hist(data['Y'], bins='auto', alpha=0.7, label='Y')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.show()

# Pairplot
sns.pairplot(data)
plt.show()

# Heatmap
plt.figure(figsize=(8, 6))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Data
X = [2, 5, 8, 15, 20, 17, 18]
Y = [3, 4, 16, 18, 21, 10, 12]

# Create a DataFrame
df = pd.DataFrame({'X': X, 'Y': Y})

# Scatterplot
scatterplot = px.scatter(df, x='X', y='Y', title='Scatterplot')
scatterplot.show()

# Boxplot
boxplot = px.box(df, y='X', title='Boxplot')
boxplot.show()

# Histogram
histogram = px.histogram(df, x='X', title='Histogram')
histogram.show()

# Pairplot
pairplot = px.scatter_matrix(df, dimensions=['X', 'Y'], title='Pairplot')
pairplot.show()

# Heatmap
heatmap = px.density_heatmap(df, x='X', y='Y', title='Heatmap')
heatmap.show()
```
